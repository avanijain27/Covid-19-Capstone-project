#### Merge US owid data with Covid Tracking national data

```python
## change date columns to datetime objects, sort by date, and reset index

US_owid_data_daily['date'] = pd.to_datetime(US_owid_data_daily.date)
national_daily['date'] = pd.to_datetime(national_daily.date)

US_owid_data_daily.sort_values(by = 'date', inplace = True)
national_daily.sort_values(by = 'date', inplace = True)

US_owid_data_daily.reset_index(drop = True, inplace = True)
national_daily.reset_index(drop = True, inplace = True)
```

```python
display(US_owid_data_daily.head(), US_owid_data_daily.tail())
```

```python
display(national_daily.head(), national_daily.tail())
```

```python
#drop rows from US owid to match date column with national
US_owid_data_daily = US_owid_data_daily.loc[(US_owid_data_daily['date'] > '2020-01-21' )&(US_owid_data_daily['date'] < '2020-10-13')].copy()
US_owid_data_daily.reset_index(drop = True, inplace = True)
```

```python
#merge US owid data and national data
US_daily = pd.merge(US_owid_data_daily, national_daily, how = "outer", on = "date")
```

```python
display(US_daily.head(), US_daily.tail())
```
